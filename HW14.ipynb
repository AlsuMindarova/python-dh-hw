# Задание 1

Необходимо написать функцию, которая считывает текст "Мира" из файла и с помощью pymystem3 лемматизирует его:

```python
from pymystem3 import Mystem
import re

def lemma(path) :
    with open(path, "r", encoding="utf8") as f:
        content = f.readlines()
    content = [x.strip() for x in content]
    content = ''.join(content)
    m = Mystem()
    lemmas = m.lemmatize(content)
    return (''.join(lemmas))

string = lemma(r'C:\Users\Алсу М\Desktop\new 1.txt')
print(string)
```
